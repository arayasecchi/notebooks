## Protein Density plots for cg_m3_pws10 runs ##

```python
norm = cm.colors.Normalize(vmax=0.2)

plt.figure( figsize = (12,12) )

for i in range(4):  
    fname=name1[i]
    plt.subplot(2,2,i+1)
    grid = gs1[i].grid
    #print (grid.shape)
    sum_vec = np.sum(grid, axis=0)
    #print (sum_vec.shape)
    #sum_vec.resize(164,177)
    rot = np.rot90(sum_vec)    
    cf = plt.contourf(rot, levels=10, origin='image', cmap = 'coolwarm', norm=norm, extend='max')
    plt.colorbar( cf )
    #plt.title(str(fname))    
    #ax.imshow(rot, norm=norm)    
        #ax.set_title(str(grid))

plt.tight_layout()  
plt.savefig("/home/raul/projects/MD-runs/synergy/SPTMRs/PRLR/CG/results/PRLR_LID_cgm3pws_prot_dens_noname.pdf", dpi=300)
plt.show()
```
